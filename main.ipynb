```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score, f1_score, matthews_corrcoef
from statistics import mean, stdev
from sklearn.metrics import confusion_matrix

df = pd.read_csv('./rtfDataSet.csv')

x = df.iloc[:, selected_features]

# Input_ y_Target_Variable.
y = df.iloc[:, -1]

# Normalize
scaler = MinMaxScaler()
# scaler.fit(self.x)
x_scaled = scaler.fit_transform(x)

criterions = ['entropy','gini']

for criterion in criterions:
    skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)
    accu_list = []
    precision_list = []
    recall_list = []
    auc_list = []
    f1_list = []
    mcc_list = []
    sensitivity_list = []
    specificity_list = []

    for train_index, test_index in skf.split(x_scaled, y):
        x_train_fold, x_test_fold = x_scaled[train_index], x_scaled[test_index]
        y_train_fold, y_test_fold = y.iloc[train_index], y.iloc[test_index]

        # Inicializar y entrenar el modelo del árbol de decisión
        model = DecisionTreeClassifier(criterion=criterion)
        model.fit(x_train_fold, y_train_fold)

        # Predecir las etiquetas para el conjunto de prueba
        y_pred = model.predict(x_test_fold)

        # Calcular métricas y agregar a las listas
        accu_list.append(accuracy_score(y_test_fold, y_pred))
        precision_list.append(precision_score(y_test_fold, y_pred))
        recall_list.append(recall_score(y_test_fold, y_pred))
        auc_list.append(roc_auc_score(y_test_fold, y_pred))
        f1_list.append(f1_score(y_test_fold, y_pred))
        mcc_list.append(matthews_corrcoef(y_test_fold, y_pred))

        # Calcular sensibilidad y especificidad
        tn, fp, fn, tp = confusion_matrix(y_test_fold, y_pred).ravel()
        sensitivity_list.append(tp / (tp + fn))
        specificity_list.append(tn / (tn + fp))

    # Imprimir resultados
    print(f"Criterio: {criterion}")
    print('Metrics Results:')
    print('-'*30)
    print(f"Accuracy: {mean(accu_list)*100:.2f}% ± {stdev(accu_list):.2f}")
    print(f"Precision: {mean(precision_list)*100:.2f}% ± {stdev(precision_list):.2f}")
    print(f"Recall: {mean(recall_list)*100:.2f}% ± {stdev(recall_list):.2f}")
    print(f"AUC: {mean(auc_list)*100:.2f}% ± {stdev(auc_list):.2f}")
    print(f"F1 Score: {mean(f1_list)*100:.2f}% ± {stdev(f1_list):.2f}")
    print(f"MCC: {mean(mcc_list)*100:.2f}% ± {stdev(mcc_list):.2f}")
    print(f"Sensitivity: {mean(sensitivity_list)*100:.2f}% ± {stdev(sensitivity_list):.2f}")
    print(f"Specificity: {mean(specificity_list)*100:.2f}% ± {stdev(specificity_list):.2f}\n")
```
